```python
import pandas as pd
import numpy as np 
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import os


if not os.path.exists("images"):
    os.mkdir("images")
```

```python
Codes = "DKK=X SEK=X"
Data = yf.download(Codes, start="2018-01-01", end="2022-08-07", interval="1d")
```

```python
Data.drop(['Adj Close', 'High', 'Low', 'Open', 'Volume'], axis=1, inplace = True) #remove unwanted columns 
```

```python
Data.columns.set_levels(['DKK/USD','SEK/USD'],level=1,inplace=True)
Data.head()
```

```python
Correlation = Data.corr()
Correlation
```

```python
Data_RO = Data.stack(level=1).reset_index(level=0).rename({'level_1': 'Ticker'}, axis=1)
Data_RO.reset_index(inplace=True)
Data_RO.rename(columns = {'index':'Ticker'}, inplace=True)
```

```python
DKKUSD = Data_RO[Data_RO["Ticker"] == 'DKK/USD']
SEKUSD = Data_RO[Data_RO["Ticker"] == 'SEK/USD']
```

```python
fig = px.line(DKKUSD, x="Date", y="Close", color = 'Ticker' , title = "Danish Kroner to USD: Correlation 0.96693")
fig.show()
fig.write_image("images/DKK.jpeg")
```

```python
fig = px.line(SEKUSD, x="Date", y="Close", color = "Ticker", color_discrete_sequence=["orange", "blue"], title = "SWEDISH Kroner to USD: Correlation 0.96693")
fig.show()
fig.write_image("images/SEK.jpeg")
```
